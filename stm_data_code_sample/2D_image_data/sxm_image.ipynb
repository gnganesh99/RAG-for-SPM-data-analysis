# Plotting STM data images

Notebook shows the working with functions to analyze STM image data files of the form .sxm

## Import modules

```python
from stm_utils import Sxm_Image
import numpy as np
import matplotlib.pyplot as plt
```

## Create file object


```python

# Load the SXM file at the specified filepath
filepath = "image.sxm"

im = Sxm_Image(filepath)
```

## Get attributes

Get the frame size of the image.

```python
frame  = im.frame
print("frame size: ", frame)
```

Get Image metadata

```python
im.header
```

# Plot images.

Get channels present in the image data file

```python
channels = im.get_channels()
print(f"Channels present: {channels}")
```

To plot the image corresponding to each channel use command: im.image(channel_name)

If channel_name is not provided, default channel is height channel: Z_Fwd

```python
image = im.image()

plt.imshow(image, origin = 'lower')
plt.colorbar()
plt.title("Z_Fwd")
plt.axis("off")
```

Get channels in the image

Plot other channels

```python
channel = "Z_Bkd"
ch_image = im.image(channel)

plt.imshow(ch_image, origin = 'lower')
plt.title(channel)
plt.colorbar()
plt.axis("off")
```

```python
channel = "Current_Fwd"
ch_image = im.image(channel)
plt.imshow(ch_image, origin='lower', cmap='viridis')
plt.title(channel)
plt.colorbar()
plt.axis("off")
```
